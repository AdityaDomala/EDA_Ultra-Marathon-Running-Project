### Goal 1  -  merge 50km and 50mi, year = 2020, event name equal to USA

```python
df2 = df[ (df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')] 
# merged 50km with 50 mi and year of event 2020 with event name USA 
```

```python
df2
```

```python
df2.tail()
```

```python
df2.shape
```

### Goal 2 - Remove (USA) from event name

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(3)
```

### Goal 3 - Athlete age data cleaning

```python
df2['athlete_age'] = df2['Year of event'] - df2['Athlete year of birth'] 
```

```python
df2['athlete_age']
```

### Goal 4 - Removing 'h' from athlete performance

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split('h').str.get(0)
```

```python
df2.head()
```

### Goal 5 - drop columns, athlette club, country, year of birth, age category

```python
df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis = 1)
```

```python
df2.head()
```

```python
df2.shape
```

### Goal 6 - Clear the null values

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

### Goal 7- check for duplicates

```python
df2[df2.duplicated() == True] # no duplicates found
```

### Goal 8 - reset the index 

```python
df2
```

```python
df2 = df2.reset_index(drop=True)
```

```python
df2
```

### Goal 9 - Changing the dtypes

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2.dtypes
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

### Goal 10 - Renaming Columns

```python
df2 = df2.rename(columns = {
    'Year of event': 'year',
    'Event dates':'race_days',
    'Event name': 'race_name',
    'Event distance/length':'race_length',
    'Event number of finishers':'race_number_of_finishers',
    'Athlete performance':'athlete_performance',
    'Athlete gender':'athlete_gender',
    'Athlete average speed':'athlete_avg_speed',
    'Athlete ID':'athlete_id',
    })
```

```python
# Seasons for the data -> Slower in summer than winter?

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12 - 2
#split between two decimals
```

```python
df2['race_month'] = df2['race_days'].str.split('.').str.get(1).astype(int)
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring')
```

```python
df2.head(15)
```

```python
df2.groupby('race_season')['athlete_avg_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```
